### Corrélation entre l'age des participants et le score de classification cross-modale obtenu:

   -  Score SMA : Classification avec les 7 regions de la SMA    
   - Score Motor: Classification avec les 76 regions motrices    
   - Train Stim: Apprentissage sur données Stimulation, Test sur imagination    
   - Train Imag: Apprentissage sur données Imagination, Test sur stimulation
   
Quand la corrélation est significative j'ai affiché la courbe de regression linéaire avec les valeurs de R2 puis p en légende
   
[Pas forcément pertinant sachant que le score de classification n'est pas "significatif" pour la majorité des sujets - a voir...]
    

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
filename='F:/IRM_marche/result\correl.csv'
from scipy.stats.stats import pearsonr
values = np.genfromtxt(filename
        , delimiter=';')
names=('AGE VS Score SMA - Train STIM ','AGE VS Score SMA - Train IMAG','AGE VS Score Motor - Train STIM','AGE VS Score Motor - Train IMAG')
col=('g','r','b','c')
for i in range(4):
    cor,p=pearsonr(values[:,i],values[:,6])
    plt.figure()
    plt.scatter(values[:, i], values[:, 6],  c=col[i])
    plt.title(names[i])
    if p<0.05:
        result=linregress(values[:, i], values[:, 6])
        X_plot = np.linspace(0.3,1,6)
        plt.plot(X_plot, X_plot*result.slope + result.intercept)
        plt.legend((cor,p))
        
```
