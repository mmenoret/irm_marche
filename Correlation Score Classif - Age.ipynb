```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
filename='F:/IRM_marche/result\correl.csv'
from scipy.stats.stats import pearsonr
values = np.genfromtxt(filename
        , delimiter=';')
names=('Score SMA - Train STIM','Score SMA - Train IMAG','Score Motor - Train STIM','Score Motor - Train IMAG')
col=('g','r','b','c')
for i in range(4):
    cor,p=pearsonr(values[:,i],values[:,6])
    plt.figure()
    plt.scatter(values[:, i], values[:, 6],  c=col[i])
    plt.title(names[i])
    if p<0.05:
        result=linregress(values[:, i], values[:, 6])
        X_plot = np.linspace(0.3,1,6)
        plt.plot(X_plot, X_plot*result.slope + result.intercept)
        plt.legend((cor,p))
        
```
